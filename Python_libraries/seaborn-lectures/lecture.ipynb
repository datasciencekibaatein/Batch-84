```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Create Dummy Data:

```python
df= pd.DataFrame({
    "Category":np.random.choice(["Electronics","Clothing","Food"],100),
    "Sales":np.random.randint(100,1000,100),
    "Profit":np.random.randint(10,300,100),
    "Region":np.random.choice(["North","South","East","West"],100)
})
```

```python
df.head()
```

```python
sns.set_style("whitegrid")
sns.set_context("paper")
```

Line plot

```python

plt.figure(figsize=(16,9))
sns.lineplot(x=df.index,y=df["Sales"])
plt.title("Sales Trend")
plt.show()
```

barplot

```python
sns.barplot(x="Category",y="Sales",data=df,estimator=np.sum)
plt.title("Avg Sales by category")
plt.show()
```

Count Plot

```python
sns.countplot(x="Region",data=df,palette="Set2")
plt.title("Category count")
plt.show()
```

Histogram

```python
plt.figure(figsize=(16,9))
sns.histplot(df["Sales"],bins=10,kde=True)
plt.title("Sales Distribution")
plt.show()
```

KDE Plot

```python
sns.kdeplot(df["Sales"],fill=True)
plt.title("Densit Sale")
plt.show()
```

Scatter Plot

```python
sns.scatterplot(x="Sales",y="Profit",data=df,hue="Category")
plt.title("Sales VS Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

Box plot

```python
sns.boxplot(x="Category",y="Sales",data=df,palette='viridis')
plt.title("Boxplot")
plt.show()
```

Violin Plot

```python
sns.violinplot(x="Region",y="Sales",data=df,palette='viridis')
plt.title("Distribution + Density")
plt.show()
```

Strip plot

```python
sns.stripplot(x="Region",y="Sales",data=df,palette="pastel")
plt.show()
```

Pair plot

```python
sns.pairplot(df,hue="Region")
```

Heatmap

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

Regplot

```python
sns.regplot(x="Sales",y="Profit",data=df)
plt.title("Regression plot")
plt.show()
```

Join Plot

```python
sns.jointplot(x="Sales",y="Profit",data=df)
```

```python
g = sns.FacetGrid(df,col="Region")
g.map(sns.histplot,"Sales")
```

```python
sns.boxplot(x='Category',y="Sales",data=df,palette='coolwarm')
sns.stripplot(x="Category",y="Sales",data=df,color='red')
```

<Strong>When to use What</strong>

|Plot | Use Case|
|----|-----|
|lineplot|Trends|
|barplot|Aggregation|
|countplot|Frequency|
|histplot|Distribution|
|scatterplot|Relationship|
|boxplot|Outliers|
|violinplot|Distribution shape|
|heatmapt|correlation|
|pairplot|EDA|
