<h1><strong>Random number generator</strong></h1>

Random numbers are extremely important in datascience , machine learning, simulations, sampling and testing algorithms.

Inside numpy library , we are having one module which is <code>random</code>

```python
import numpy as np
```

<code>rand()</code>

```python
np.random.rand()
```

```python
print(np.random.rand(5))
```

```python
np.random.rand(2,3)
```

<code>randint()</code>

```python
np.random.randint(1,7,(2,3))
```

<code>choice()</code>

```python
np.random.choice([34,78,92,57,91,56],10)
```

```python
np.random.choice([23,45,12,78],size=10,p=[0.1,0.2,0.3,0.4])
```

<code>shuffle()</code>

```python
arr = np.array([1,2,4,5,6,7])
np.random.shuffle(arr)
print(arr)
```

<code>permutation()</code>

```python
arr = np.array([1,2,3,4,5,6])
print(np.random.permutation(arr))
print(arr)
```

<h2 style="color:red;"><strong>seed()</strong></h2>

Sometimes we want same random numbers every time (for reproducibility)

```python
np.random.seed(42)
np.random.rand(3)
```

<h3>Random numbers from <code>normal distribution</code></h3>

```python
np.random.randn(3,3)
```

<h3>Generate normal distribution with <code>Mean and Std</code></h3>

```python
np.random.normal(loc=50,scale=10,size=4)
```

<h3>Uniform Distribution</h3>

Generate numbers uniformly between two values.

```python
np.random.uniform(low=10,high=20,size=5)
```

<h3>Binomial Distribution</h3>

Used for success / failure experiments

```python
np.random.binomial(n=10,p=0.999,size=5)
```

<h3>Poisson Distribution</h3>

Used for even frequency

- Website visits
- Number of calls in call center
- Traffic flow

```python
np.random.poisson(lam=5,size=6)
```

<h2>Generate one dataset</h2>

```python
np.random.seed(42)

rating= np.random.rand(100)
income = np.random.randint(25000,88000,100)
age = np.random.randint(18,58,100)
target = np.random.randint(0,2,100)
city = np.random.choice(["mumbai","delhi","chandigarh","banglore","lucknow","noida"],100)

import pandas as pd

df = pd.DataFrame(rating,columns=['rating'])
df['age'] = age
df['city'] = city
df['income'] = income
df['target'] = target

df.head(10)
```
