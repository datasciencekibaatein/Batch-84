<h1>Label Encoding</h1>

Assigns each category with a unique integer.

```python
from sklearn.preprocessing import LabelEncoder

df['City']

le = LabelEncoder()
encoded = le.fit_transform(df['City'])
```

```python
encoded
```

<h1>One-Hot Encoding</h1>

Creates binary columns for each category

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)

encoded = ohe.fit_transform(df[['Payment_Method']])
encoded
```

<h1>Ordinal Encoding</h1>

Used when categories have a meaningful order

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[["Low","Medium","High"]])

encoded = oe.fit_transform(df[['Satisfaction_Level']])
encoded
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder

education_order=['High School','Bachelor','Master','PhD']
satisfaction_order= ['Low','Medium','High']

le = LabelEncoder()
oe = OrdinalEncoder(categories=[education_order,satisfaction_order])

df['Subscribed'] = le.fit_transform(df['Subscribed'])
df['Purchased'] = le.fit_transform(df['Purchased'])
df[['Education_Level','Satisfaction_Level']] = oe.fit_transform(df[['Education_Level','Satisfaction_Level']])
```

```python
df = pd.get_dummies(df,columns=['City','Gender','Payment_Method','Product_Category'])
```
