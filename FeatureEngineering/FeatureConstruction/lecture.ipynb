```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 100

data = pd.DataFrame({
    'age':np.random.randint(18,60,n),
    "salary":np.random.randint(20000,100000,n),
    "ciyt":np.random.choice(['Delhi','Mumbai','Pune'],n),
    "gender":np.random.choice(['Male','Female'],n),
    "purchased":np.random.choice([0,1],n)
})

data.head()
```

<h2>Feature Construction</h2>
Feature construction is the process of creating new features from existing data to improve model performance.

```python
data['salary_per_age'] = data['salary']/data['age']
```

```python
data.head()
```

```python
data['Age_group'] = pd.cut(data['age'],bins=[18,30,45,60],labels=['Young','Mid','Senior'])
```

<h2>Feature Scaling</h2>

Transforming features so they are on similar scale

- Distance based models it is necessary
- Prevent feature dominance

StandardScaler(z-score)

z = (x - mean)/std

Inverse = (z x std + mean)

```python
data['age'].mean()
```

```python
data['age'].std()
```

```python
data['manual_scaled']=(data['age']-data['age'].mean())/data['age'].std()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data[['Age_scaled','salary_scaled']] = scaler.fit_transform(data[['age','salary']])
```

```python
data.head()
```

X = (x-min) / (max - min)

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[['salary_scaled_min_max']] = scaler.fit_transform(data[['salary']])
```

```python
data['manual_min_max_salary'] = (data['salary'] - data['salary'].min())/(data['salary'].max()-data['salary'].min())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(data[['age','salary']])

data[['age_scaled','salary_scaled']] = scaled
```

```python

original = scaler.inverse_transform(scaled)

data[['age_original','salary_original']] = original
```
