```python


import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing


housing = fetch_california_housing(as_frame=True)

df = housing.frame.copy()

print("Original Shape:", df.shape)
print(df.head())


np.random.seed(42)

cols_to_miss = ['MedInc', 'HouseAge', 'AveRooms', 'Population']

for col in cols_to_miss:
    df.loc[df.sample(frac=0.10).index, col] = np.nan

df.loc[df.sample(frac=0.20).index, 'AveOccup'] = np.nan

df.loc[df['Latitude'] > 36, 'Longitude'] = np.nan

print("\nMissing Values:")
print(df.isnull().sum())



df.to_csv("missing_value_practice_data.csv", index=False)

print("\nDataset saved as missing_value_practice_data.csv")
print(df.head())

```

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("missing_value_practice_data.csv")
```

```python
df = pd.DataFrame(data)
```

```python
df.isnull().sum()
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled,columns=df.columns)
df_scaled
```

```python
imputer = KNNImputer(n_neighbors=3)
```

```python
df_imputed = imputer.fit_transform(df_scaled)
```

```python
final_df = pd.DataFrame(scaler.inverse_transform(df_imputed),columns=df.columns)
```
