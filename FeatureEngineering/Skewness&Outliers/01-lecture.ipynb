<h1><strong>Outliers</strong></h1>


An outlier is a data point that differs significantly from other obervations in a dataset. It is a value that lies at an abnormal distance from other values in a random sample of a population.

- Global outliers (Point Anomalies)
- Contextual Outliers (conditional Anomalies)
- Collective outliers (Group anomalies)

<h1><strong>Skewness</strong></h1>

Skewness measures the asymmetry of a probability distribution. A perfectly symmetrical distribution (like normal bell curve) has skewness = 0. When data is skewed, the tail on one side is longer or heavier than other.

- Left Skew(Negative)
- Symmtric(normal)
- Right Skew(Postive)
