<h1><strong>Skewness</strong></h1>

Skewness is measure of the asymmetry of the probability distribution.

- Symmetric: Mean = Median = Mode
- Right Skewed: Mean > Median > Mode (Tail pulls to right)
- Left skewed: Mean< Median < Mode (Tail pulls to left)

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sym_data = np.random.normal(0,1,1000)
sns.histplot(sym_data,kde=True)
plt.title("Symmetric distribution")
plt.show()
```

```python
right_skewed = np.random.exponential(scale=2,size=1000)
sns.histplot(right_skewed,kde=True)
plt.title("Right Skewed")
plt.show()
```

```python
left_skewed = 100-np.random.exponential(scale=2,size=1000)
sns.histplot(left_skewed,kde=True)
plt.title("left skewed")
plt.show()
```

```python
#Right Skewed / Log transformation

log_data = np.log1p(left_skewed)

log_data
```

```python
right_skewed = np.random.exponential(scale=2,size=1000)
sns.histplot(log_data,kde=True)
plt.title("Right Skewed")
plt.show()
```

```python
sqrt_data = np.sqrt(left_skewed)
```

```python
right_skewed = np.random.exponential(scale=2,size=1000)
sns.histplot(sqrt_data,kde=True)
plt.title("Right Skewed")
plt.show()
```

```python
square_data = np.square(left_skewed)
```

```python
right_skewed = np.random.exponential(scale=2,size=1000)
sns.histplot(square_data,kde=True)
plt.title("Left Skewed")
plt.show()
```
