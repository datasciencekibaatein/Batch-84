<h2>Operations</h2>

```python
ar = np.array([3,4,5,3,23])
```

Operation with scaler value

```python
an = ar * 3
an = ar / 2
an = ar // 2
an = ar % 2
print(an)
```

```python
mat = np.array(
    [
        [2,7,5],
        [6,8,0],
        [2,5,4]
    ]
)
```

```python
print(mat*3)
```

```python
ar1 = np.array([4,3,5])
ar2 = np.array([6,8,2])

print(ar1 + ar2)
```

```python
mat1 = np.array(
    [
    [3,4,2],
    [5,6,2],
    [5,6,2]
    ]
)

mat2 = np.array([
    [3,5,1],
    [5,7,2],
    [3,2,1]
])
```

```python
mat3 = np.array(
    [
        [1,2,3],
        [2,1,1],
        [2,3,4]
    ]
)

mat4 = np.array(
    [
        [1,3,1],
        [3,2,2],
        [2,1,1]
    ]
)
```

```python
mat3@mat4
```
