# Silhouette Score

<h3>Why do we need silhouette score?</h3>

Suppose K-Means produced 3 clusters.

The algorithm will always create clusters, even if they are poor.

- Are these clusters actually good?

Silhouette score helps us measures that quantitatively.

<h3>Intuition</h3>

a = Intra-Cluster distance

- Average distance frmo the point to all other points in the same cluster.


b = Inter-Cluster Distance

- Average distance from the point to points in the nearest neighboring cluster.

<h3>Range of silhouette score</h3>

-1 <= S <= 1

|score | Meaning|
|----|-----|
|0.71 to 1.0 | Excellent clustering|
|0.51 to 0.70 | Good clustering|
|0.26 to 0.50 | Weak structure|
|0.00 to 0.25 | No significat structure|
|<0|Wrong clusterin|

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
X,_ = make_blobs(
    n_samples =500,
    centers=4,
    random_state=42
)

```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=_)
plt.show()
```

```python
kmeans = KMeans(
    n_clusters=6,
    random_state=42
)
```

```python
labels = kmeans.fit_predict(X)
```

```python
labels
```

```python
score = silhouette_score(X,labels)
```

```python
print(score)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=labels)
plt.show()
```
