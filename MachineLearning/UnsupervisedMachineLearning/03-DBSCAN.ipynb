# DBSCAN(Density Based Spatial Clustering)

DBSCAN is one of the most powerful clustering algorithm because:

- It does not require specifying the number of clustering (K) beforehand.
- It can find clusters of arbitrary shapes.
- it can identify outliers/noise automatically.
- Works well when clusters are separated by low-density regions.

<h3>Core Idea</h3>

DBSCAN creates clusters based on density.

It requires only two parameters

- Epsilon(Radius around a point.)
- MinPts(Minimum number of points required inside radius(epsilon))

<h3>Types of points in DBSCAN</h3>

There are 3 types.

<h4>Core point</h4>

A point having at least MinPts neighbors within epsilon.

<h4>Border point</h4>

Not enough neighbors itself, but lies near a core point.

<h4>Noise Point</h4>

Point far from any dense resion.

<h3>DBSCAN Parameters</h3>

eps: Neighborhood radius.

Small eps:

- Many small clusters
- More Noise


Large eps:

- Clusters merge
- less noise


min_samples: Minimum points required

small value:

- Many clusters
- Sensitive to noise

Large value:

- Stricter clustering


```python
from sklearn.datasets import make_moons

import matplotlib.pyplot as plt

X,y = make_moons(
    n_samples = 500,
    noise = 0.08,
    random_state=42
)

```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(
    eps=0.1,
    min_samples=5
)

clusters = dbscan.fit_predict(X)

print(clusters)
```

```python
plt.scatter(X[:,0],X[:,1],c=clusters,cmap='rainbow')
plt.show()
```
