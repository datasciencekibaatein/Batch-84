## K-Means Clustering

K-Means is an unsupervised ml algorithm that groups similar data points into K clusters.

#### The algorithm tries to:

- Minimize distance within a cluster
- Maximize distance between clusters

#### The word:

- K = Number of clusters
- Means = Cluster center (average value)

### Algorithms

- Step 01: Initialize K centroids randomly.
- Step 02: Assign every point to nearest centroid.
- Step 03: Update centroid position.
- Step 04: Repeat steps 2 and 3 until convergence.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
iris = load_iris()

X = iris.data
print(X.shape)
```

```python
iris.target_names
```

```python
df = pd.DataFrame(X,columns=iris.feature_names)
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X[:,0],
    X[:,1]
)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
k = 3
```

```python
kmeans = KMeans(n_clusters=3,random_state=42)

kmeans.fit(X)
```

```python
iris.target
```

```python
labels = kmeans.labels_

print(labels)
```

```python
centroids = kmeans.cluster_centers_

print(centroids)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X[:,0],
    X[:,1],
    c=labels,
    cmap = 'viridis'
)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.title("K-Means Clusters")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X[:,0],
    X[:,1],
    c=labels,
    cmap = 'viridis'
)

plt.scatter(
    centroids[:,0],
    centroids[:,1],
    s=300,
    c='red',
    marker='x'
)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.title("Clusters with centroids")

plt.show()
```

### WCSS(Within clusters sum of squared)



```python
kmeans.inertia_
```

```python
inertia = []

for k in range(1,11):
    model = KMeans(
        n_clusters=k,
        random_state=42
    )
    model.fit(X)
    inertia.append(model.inertia_)
```

```python
inertia
```

```python
plt.figure(figsize=(8,6))

plt.plot(
    range(1,11),
    inertia,
    marker='o'
)

plt.xlabel("Number of clusters (K)")
plt.ylabel("Inertia")

plt.title("Elbow method")

plt.show()
```
