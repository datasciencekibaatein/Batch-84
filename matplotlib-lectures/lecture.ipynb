```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Line Chart

```python
x = np.arange(1,6)
y = [10,20,15,25,17]
```

```python
plt.plot(x,y,)
plt.title("Line chart")
plt.xlabel("X values")
plt.ylabel("y values")
plt.show()
```

Multiple line plot

```python
y1 = [20,12,16,21,18]
y2 = [4,8,10,15,19]
```

```python
plt.plot(x,y,label="Product a")
plt.plot(x,y1,label="Product b")
plt.plot(x,y2,label="Product c")
plt.legend()
plt.title("Mulitple line chart")
plt.xlabel("X values")
plt.ylabel("y values")
plt.show()
```

Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y,c="red",marker="x",alpha=0.5)
plt.title("Scatter plot")
plt.show()
```

Bar Chart

```python
categories = ['A','B','C','D']
values = [10,25,15,28]
```

```python
plt.bar(categories,values,width=0.4,color="cyan")
plt.title("Bar chart")
plt.show()
```

Horizontal  bar chart

```python
plt.barh(categories,values)
```

Histogram

```python
data = np.random.randn(1000)
plt.hist(data,bins=30,histtype="bar")
plt.show()
```

Pie Chart

```python
labels = ['a','b','c']
sizes = [40,35,25]

plt.pie(sizes,labels=labels,autopct="%1.1f%%")
plt.show()
```

Box Plot

```python
data = [np.random.randn(100),np.random.randn(100)]
plt.boxplot(data)
plt.show()
```

Area plot

```python
x = np.arange(1,6)
plt.stackplot(x,y1,y2)
```

```python
x = np.arange(1,6)
y = [10,20,15,23,28]

plt.figure(figsize=(16,9))

plt.subplot(1,2,1)
plt.plot(x,y,linestyle='--',marker="o")
plt.title("line chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("bar chart")

plt.show()
```
