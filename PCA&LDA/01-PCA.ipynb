# PCA (Principal Component Analysis)

Principal Component analysis is one of the most important dimensionality reduction techniques in Machine Learning.

It helps us:

- Reduce the number of features
- Remove the redundancy
- Visualize high-dimension data
- speed up model training
- Reduce overfitting

## Why Do We need PCA?

Cols : Student, Height, Weight, ShoeSize

Notice: 

- Taller students generally have larger shoe sizes.
- Height and shoe size are highly correlated.

Therefore both features are carrying almost the same information.

PCA helps us to combine such features into a smaller set of meaningful features while preserving maximum information.

### Terminology

##### Variance : 
Variance tells how spread out data is.

Dataset A = [10,10,10,10,10]
variance = 0 

Dataset B = [5,10,15,30,45]
Variance = High

More variance means more information

PCA tries to preserve directions with maximum Variance

##### Covariance :
Covariance tells how two variables move together.

Positive Covariance

Height(high)
Weight (high)


Negative Covariance

Temperature(high)
Sweater sales (low)


PCA uses covariance to find relationships between features.


##### Eigenvectors and Eigenvalues


Eignevectors:

    Direction of maximum variance.

Eigenvalue:

    Amount of variance captured in that direction
