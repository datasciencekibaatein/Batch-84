# Linear Discriminant Analysis(LDA)

<strong style="color:red;">LDA is for supervised</strong>

LDA is supervised dimensionality reduction and classification technique that transforms data into a lower-dimensional space while maximizing the separation between different classes.

```python
from sklearn.datasets import load_iris
iris = load_iris()

X = iris.data
y = iris.target
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
import pandas as pd


pd.DataFrame(X_scaled).head()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(
    n_components = 2
)
```

```python
x_lda = lda.fit_transform(X_scaled,y)
```

```python
x_lda.shape
```

```python
import matplotlib.pyplot as plt

plt.scatter(
    x_lda[:,0],
    x_lda[:,1],
    c= y
)

plt.xlabel("LD1")
plt.ylabel("LD2")
plt.show()
```
