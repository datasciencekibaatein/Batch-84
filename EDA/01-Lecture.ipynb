<h2>Merge all sheets</h2>

```python
df = employees.merge(departments,on='DeptID',how='left') \
              .merge(locations,on='LocationID',how='left') \
              .merge(jobs,on='JobID',how='left') \
              .merge(attendance,on='EmployeeID',how='left') \
              .merge(attrition,on='EmployeeID',how='left')
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum() / len(df) * 100
```

```python
df['ExitReason'] = df['ExitReason'].fillna('Active')
df['NoticePeriodDays'] = df['NoticePeriodDays'].fillna(0)
```

```python
len(df[df['ExitReason']!='Active'])/len(df)*100
```

<h2>Key Metrics</h2>

```python
metrics = {
    'Total Employees': len(df),
    'Average Salary' : df['Salary'].mean(),
    'Attrition Rate (%)' : len(df[df['ExitReason']!='Active'])/len(df)*100,
    'Avg Engagement Score': df['EngagementScore'].mean(),
    'Avg Satisfaction Score': df['SatisfactionScore'].mean(),
    'Total Bonus Pool' : df['Bonus'].sum(),
    'Avg Training hours' : df['TrainingHours'].mean(),
    'Avg Age': df['Age'].mean(),
    'Total Departments': df['Department'].nunique(),
    'Avg year since Promotion': df['YearsSincePromotion'].mean()   
    }
```

```python
for k,v in metrics.items():
    print(f"{k}: {v:,.2f}")
```
